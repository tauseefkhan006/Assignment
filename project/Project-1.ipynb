```python
plt.figure(figsize = (15,8))
sns.barplot( data=df, x= "matchType" ,y = "killPoints" ,ci = None )
plt.xticks(rotation = 70)
plt.title(" matchType vs killPoints")
plt.show()

```

from the bar plot we can see that the kill points of normal-duo-fpp have the 800-900. 
so most of the players playing Duo fpp matchs thats is why they get the maximum kill points.

```python
plt.figure(figsize = (15,5))
sns.barplot( data=df, x= "matchType" ,y = "weaponsAcquired" ,ci = None , color = "b")
plt.xticks(rotation = 70)
plt.title(" matchType vs weaponsAcquired")
plt.show()
```

 in normal-solo-fpp matchType most wepons acquired by player, it also depends servival duration .
